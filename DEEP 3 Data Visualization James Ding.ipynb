```python
# import libraries 
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

tp = pd.read_csv('texas_affordability_census_tract.csv')
tp["cbsa"] = tp["cbsa"].str[1:-1]
urban = ["San Antonio-New Braunfels, TX", "Houston-The Woodlands-Sugar Land, TX", "Austin-Round Rock, TX", "Dallas-Fort Worth-Arlington, TX"]
for cities in urban:
    tpu = tp.loc[tp.cbsa == cities]
#tpu = tp[tp.cbsa in urban]
```

```python
#Lower cost of transportation = higher employment empowerment (urban areas only)
graph = sns.scatterplot(x="t_cost_ami", y="emp_ovrll_ndx", data=tpu) #cost of transportation vs access to employment (urban only)
graph.set(xlabel="Cost of Transportation", ylabel = "Access to Employment")
#Hypothesis on cost of transportation vs access to employment is satisfied
```

```python
#Lower cost of transportation = higher employment empowerment
graph = sns.scatterplot(x="t_cost_ami", y="emp_ovrll_ndx", data=tp) #cost of transportation vs access to employment 
graph.set(xlabel="Cost of Transportation", ylabel = "Access to Employment")
#Hypothesis is not satisfied
```

```python
#Higher cost of transportation leads to higher auto ownership cost (urban only)
graph = sns.scatterplot(x="t_cost_ami", y="auto_ownership_cost_ami", data=tpu) #cost of transportation vs auto ownership cost(urban only)
graph.set(xlabel="Cost of Transportation", ylabel = "Autownership cost")
#Hypothesis is satisfied (auto ownership is part of transportation)
```

```python
#Lower cost of transportation = better employment access
graph = sns.scatterplot(x="t_cost_ami", y="emp_gravity", data=tpu) #cost of transportation vs employment access
graph.set(xlabel="Cost of Transportation", ylabel = "Employment potential")
#Hypotehsis is satisfied
```

```python
#Hypothesis: The more compact the neighborhood, the cheaper the transportation
graph = sns.scatterplot(x="t_cost_ami", y="compact_ndx", hue = "gross_hh_density", data=tpu) #cost of transportation vs compact of neighborhood
graph.set(xlabel="Cost of Transportation", ylabel = "Compactness of Neighborhood")
#Hypothesis is satisfied (the more compact, the more dense is the neighborhood)
```

```python
#Hypothesis: The cheaper the trasnportation, the higher percentage of people will take public transportation
graph = sns.scatterplot(x="t_cost_ami", y="h_cost", data=tpu) 
graph.set(xlabel="Cost of Transportation", ylabel = "Percentage of People taking Public Transit")
#Hypothesis is not supported
```

```python
#the cheaper the transportation, the fewer cars each household will have
graph = sns.scatterplot(y = "autos_per_hh_ami", x = "t_cost_ami", hue = "co2_per_hh_local", data = tpu) #cost of transportation vs number of cars owned
graph.set(xlabel="Cost of Transportation", ylabel = "Auto ownership per household")
#Hypthesis is satisfied. Furthermore, carbon emission per household increases as car number goes up
```

```python
#cost of transportation vs total miles vehicle travels
graph = sns.scatterplot(y = "vmt_per_hh_ami", x = "t_cost_ami", hue = "co2_per_hh_local", data = tpu)
graph.set(xlabel="Cost of Transportation", ylabel = "Vehicle per household")
#Hypothesis is satisfied
```
